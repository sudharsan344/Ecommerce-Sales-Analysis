```python
sales_category=df.groupby("Category")["Sales"].sum()
sales_category.plot(kind="bar",figsize=(8,5))

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
profit_category=df.groupby("Category")["Profit"].sum()
profit_category.plot(kind="bar",color="darkblue",figsize=(8,5))


plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
top_products=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar",figsize=(8,5))

plt.title("Top 10 product by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
Region_Sales=df.groupby("Region")["Sales"].sum()
Region_Sales.plot(kind="bar",figsize=(8,5))

plt.title("Region By Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
Segment_sales = df.groupby("Segment")["Sales"].sum()
Segment_sales.plot(kind="pie",autopct="%1.1f%%")

plt.title("Sales by segment")
plt.ylabel("")

plt.show()
```
